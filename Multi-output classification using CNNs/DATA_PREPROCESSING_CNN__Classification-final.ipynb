```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv1D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay

# Load sensor data files (assuming tab-delimited files)
sensor_files = ['PS1.csv', 'PS2.csv', 'PS3.csv', 'PS4.csv', 'PS5.csv', 'PS6.csv', 'EPS1.csv',
                'FS1.csv', 'FS2.csv', 'TS1.csv', 'TS2.csv', 'TS3.csv', 'TS4.csv', 'VS1.csv',
                'CE.csv', 'CP.csv', 'SE.csv']

# Load the 17 sensor data files
sensor_data = []
for file in sensor_files:
    data = pd.read_csv(file, sep=',', header=None)
    sensor_data.append(data)

# Check if all data arrays have the same shape
shapes = [data.shape for data in sensor_data]
print("Shapes of all sensor data:", shapes)

# Split data by frequency
X_100Hz = np.array(sensor_data[:7])
X_10Hz = np.array(sensor_data[7:9])
X_1Hz = np.array(sensor_data[9:])

# Resample the data
def resample_data(data, factor):
    length = data.shape[1]
    if length % factor != 0:
        data = data[:, :length - (length % factor)]
    return data.reshape(data.shape[0], -1, factor).mean(axis=2)

X_100Hz_resampled = np.array([resample_data(cycle, 100) for cycle in X_100Hz])
X_10Hz_resampled = np.array([resample_data(cycle, 10) for cycle in X_10Hz])
X_1Hz_resampled = np.array(X_1Hz)

# Padding or truncating data to have the same number of sensors
def pad_data(data, target_shape):
    num_sensors = target_shape[0]
    current_shape = data.shape[0]
    if current_shape < num_sensors:
        pad_width = ((0, num_sensors - current_shape), (0, 0), (0, 0))
        return np.pad(data, pad_width, mode='constant', constant_values=0)
    elif current_shape > num_sensors:
        return data[:num_sensors]
    return data

desired_num_sensors = 17

X_100Hz_padded = pad_data(X_100Hz_resampled, (7, X_100Hz_resampled.shape[1], X_100Hz_resampled.shape[2]))
X_10Hz_padded = pad_data(X_10Hz_resampled, (2, X_10Hz_resampled.shape[1], X_10Hz_resampled.shape[2]))
X_1Hz_padded = pad_data(X_1Hz_resampled, (8, X_1Hz_resampled.shape[1], X_1Hz_resampled.shape[2]))

X_combined = np.concatenate([X_100Hz_padded, X_10Hz_padded, X_1Hz_padded], axis=0)

target_data = pd.read_csv('profile.csv', sep=',', header=None)
y = target_data.values

target_1_mapping = {3: 0, 20: 1, 100: 2}
target_2_mapping = {73: 0, 80: 1, 90: 2, 100: 3}
target_3_mapping = {0: 0, 1: 1, 2: 2}
target_4_mapping = {90: 0, 100: 1, 115: 2, 130: 3}
target_5_mapping = {0: 0, 1: 1}

y1_mapped = np.array([target_1_mapping[val] for val in y[:, 0]])
y2_mapped = np.array([target_2_mapping[val] for val in y[:, 1]])
y3_mapped = np.array([target_3_mapping[val] for val in y[:, 2]])
y4_mapped = np.array([target_4_mapping[val] for val in y[:, 3]])
y5_mapped = np.array([target_5_mapping[val] for val in y[:, 4]])

y1 = to_categorical(y1_mapped, num_classes=3)
y2 = to_categorical(y2_mapped, num_classes=4)
y3 = to_categorical(y3_mapped, num_classes=3)
y4 = to_categorical(y4_mapped, num_classes=4)
y5 = to_categorical(y5_mapped, num_classes=2)

y_combined = np.concatenate([y1, y2, y3, y4, y5], axis=1)

X_train, X_test, y_train, y_test = train_test_split(X_combined.transpose(1, 2, 0), y_combined, test_size=0.2, random_state=42)

scaler_X = StandardScaler()
X_train = scaler_X.fit_transform(X_train.reshape(-1, X_train.shape[-1])).reshape(X_train.shape)
X_test = scaler_X.transform(X_test.reshape(-1, X_test.shape[-1])).reshape(X_test.shape)

# Define the model
input_layer = Input(shape=(60, 17))
x = Conv1D(128, kernel_size=5, activation='relu')(input_layer)
x = BatchNormalization()(x)
x = Dropout(0.3)(x)

x = Conv1D(64, kernel_size=3, activation='relu')(x)
x = BatchNormalization()(x)
x = Dropout(0.3)(x)

x = Flatten()(x)

# Define outputs for each target
output1 = Dense(3, activation='softmax')(x)
output2 = Dense(4, activation='softmax')(x)
output3 = Dense(3, activation='softmax')(x)
output4 = Dense(4, activation='softmax')(x)
output5 = Dense(2, activation='softmax')(x)

# Create the model
model = Model(inputs=input_layer, outputs=[output1, output2, output3, output4, output5])

# Compile the model with a lower learning rate
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(X_train, [y_train[:, :3], y_train[:, 3:7], y_train[:, 7:10], y_train[:, 10:14], y_train[:, 14:]],
                    epochs=100,
                    batch_size=32,
                    validation_split=0.2)

# Evaluate the model
losses_and_accuracies = model.evaluate(X_test, [y_test[:, :3], y_test[:, 3:7], y_test[:, 7:10], y_test[:, 10:14], y_test[:, 14:]])
print('Losses and accuracies for each output:')
for i, loss_or_accuracy in enumerate(losses_and_accuracies):
    print(f'Output {i+1}: {loss_or_accuracy}')

# Generate classification reports for each target
y_pred = model.predict(X_test)
for i, (y_t, y_p) in enumerate(zip([y_test[:, :3], y_test[:, 3:7], y_test[:, 7:10], y_test[:, 10:14], y_test[:, 14:]],
                                   y_pred)):
    print(f'Classification Report for Output {i+1}:')
    print(classification_report(y_t.argmax(axis=1), y_p.argmax(axis=1)))

# Confusion Matrix and Plotting
for i, (y_t, y_p) in enumerate(zip([y_test[:, :3], y_test[:, 3:7], y_test[:, 7:10], y_test[:, 10:14], y_test[:, 14:]],
                                   y_pred)):
    y_true = y_t.argmax(axis=1)
    y_pred_classes = y_p.argmax(axis=1)
    
    cm = confusion_matrix(y_true, y_pred_classes)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=np.arange(cm.shape[0]))
    
    plt.figure(figsize=(8, 6))
    disp.plot(cmap=plt.cm.Blues)
    plt.title(f'Confusion Matrix for Output {i + 1}')
    plt.show()


```

```python
# Print all keys in the history to check their names
print("History keys:", history.history.keys())

import matplotlib.pyplot as plt

# Plot accuracy vs epochs
plt.figure(figsize=(12, 6))

# Plot accuracies for each output
for j in range(5):
    plt.plot(history.history[f'dense_{ 5+ j}_accuracy'], label=f'Training Accuracy Output {j + 1}')
    plt.plot(history.history[f'val_dense_{5 + j}_accuracy'], label=f'Validation Accuracy Output {j + 1}', linestyle='--')

plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epochs')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```
